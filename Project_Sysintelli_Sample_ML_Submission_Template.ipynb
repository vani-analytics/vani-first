## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothesis 1: Active Membership vs Customer Churn

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Statement: Active members are less likely to leave the bank than inactive members.
Null Hypothesis (H₀)

There is no association between Active Membership (IsActiveMember) and Customer Churn (Exited).

Alternative Hypothesis (H₁)

There is a significant association between Active Membership (IsActiveMember) and Customer Churn (Exited).

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(
    df['IsActiveMember'],
    df['Exited']
)

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("P-value:", p)

if p < 0.05:
    print("Reject H0")
    print("Active Membership significantly affects Customer Churn")
else:
    print("Fail to Reject H0")
````

##### Which statistical test have you done to obtain P-Value?

Chi-Square Test of Independence

##### Why did you choose the specific statistical test?

Because:

IsActiveMember → Categorical (0,1)
Exited → Categorical (0,1)

Chi-Square is used to test the relationship between two categorical variables.

### Hypothesis 2: Customer Age Influences Churn

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Statement

Customers who leave the bank are older than customers who stay.

Null Hypothesis (H₀)

Average age of churned and non-churned customers is equal.

Alternative Hypothesis (H₁)

Average age differs significantly between churned and non-churned customers.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import ttest_ind

churned = df[df['Exited']==1]['Age']

retained = df[df['Exited']==0]['Age']

t_stat, p = ttest_ind(
    churned,
    retained
)

print("T Statistic:", t_stat)
print("P-value:", p)

if p < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Independent T-Test

##### Why did you choose the specific statistical test?

Because:

Age → Numerical variable
Exited → Two independent groups

The Independent T-Test compares means between two groups.

### Outlier Detection Using IQR

````python
for col in numerical_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower) |
        (df[col] > upper)
    ]

    print(col, ":", len(outliers))
````

### 5. Data Transformation

#### Do you think that your data needs to be transformed? If yes, which transformation have you used. Explain Why?

Yes, data transformation was required before training the machine learning models. Machine learning algorithms generally perform better when numerical features are on a similar scale. In this dataset, features such as Balance, EstimatedSalary, CreditScore, and Age have different ranges and units.

Therefore, Standardization (Standard Scaling) was applied to transform the numerical features.

Why Standardization?
It scales features to have a mean of 0 and a standard deviation of 1.
Prevents features with larger values from dominating the model.
Improves the performance of distance-based and optimization-based algorithms such as Logistic Regression, KNN, and SVM.
Helps the model converge faster during training.

The formula used for Standardization is:

Z=
(X−μ)/σ
	​


where:

X = Original value
μ = Mean
σ = Standard deviation

````python
from sklearn.preprocessing import StandardScaler

# Define Features and Target
X = df.drop('Exited', axis=1)
y = df['Exited']

# Numerical columns to scale
num_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

# Apply Standard Scaling
scaler = StandardScaler()

X[num_cols] = scaler.fit_transform(X[num_cols])

# Check transformed data
X.head()# Transform Your data
````

Mean = 0
Standard Deviation = 1

The transformation formula is:

Z=
(X−μ)/σ
	​


where:

X = Original value
μ = Mean of the feature
σ = Standard deviation of the feature
Why Standard Scaling?
The dataset contains numerical features with different ranges, such as:
Age
Balance
CreditScore
EstimatedSalary
Features with larger values can dominate the model if scaling is not applied.
Standard Scaling improves the performance of machine learning algorithms such as:
Logistic Regression
Support Vector Machine (SVM)
K-Nearest Neighbors (KNN)
It helps the model converge faster and improves prediction accuracy.

Therefore, Standard Scaling was applied to ensure that all numerical features contribute equally during model training and to improve overall model performance.

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- Start of Fix: Re-apply full preprocessing pipeline ---
# The error "ValueError: could not convert string to float" indicates
# that categorical (string) columns were passed to LogisticRegression's fit method.
# We need to ensure X_train and X_test are fully numerical.

# Start with a clean copy of the original DataFrame from the kernel state
df_temp = df.copy()

# 1. Drop non-informative columns (as identified in the notebook)
df_temp = df_temp.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)

# 2. Create New Feature: Balance per Product (as in the notebook)
df_temp['BalancePerProduct'] = df_temp['Balance'] / (df_temp['NumOfProducts'] + 1)

# 3. Create New Feature: Age Group (as in the notebook)
df_temp['AgeGroup'] = pd.cut(
    df_temp['Age'],
    bins=[18, 30, 45, 60, 100],
    labels=['Young', 'Adult', 'Middle_Aged', 'Senior'],
    right=False # To include the lower bound in the bin
)

# 4. One-hot encode categorical columns (Geography, Gender, AgeGroup)
# These were not encoded in the X_train/X_test that caused the error.
df_temp = pd.get_dummies(df_temp, columns=['Geography', 'Gender', 'AgeGroup'], drop_first=True)

# 5. Separate features (X) and target (y)
X_final = df_temp.drop('Exited', axis=1)
y_final = df_temp['Exited']

# 6. Data Splitting (re-doing the split with the correctly processed X_final)
X_train, X_test, y_train, y_test = train_test_split(
    X_final,
    y_final,
    test_size=0.20,
    random_state=42,
    stratify=y_final
)

# 7. Apply Standard Scaling to numerical features
# Identify numerical columns in the processed X_final that need scaling.
# Exclude binary one-hot encoded columns from this list as they are already 0/1.
numerical_cols_to_scale = [
    'CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts',
    'EstimatedSalary', 'BalancePerProduct'
]

scaler = StandardScaler()

# Fit and transform scaler on X_train for selected numerical columns
X_train[numerical_cols_to_scale] = scaler.fit_transform(X_train[numerical_cols_to_scale])

# Transform X_test using the scaler fitted on X_train
X_test[numerical_cols_to_scale] = scaler.transform(X_test[numerical_cols_to_scale])
# --- End of Fix ---

# Train Model
lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)

# Predictions
y_pred = lr.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report")
print(classification_report(y_test, y_pred))
# ML Model - 1 Implementation

# Fit the Algorithm

# Predict on the model
````
